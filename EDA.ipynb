```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
flair = nib.load('./data/BraTS2021_00000/BraTS2021_00000_flair.nii.gz')
seg = nib.load('./data/BraTS2021_00000/BraTS2021_00000_seg.nii.gz')
t1 = nib.load('./data/BraTS2021_00000/BraTS2021_00000_t1.nii.gz')
t1ce = nib.load('./data/BraTS2021_00000/BraTS2021_00000_t1ce.nii.gz')
t2 = nib.load('./data/BraTS2021_00000/BraTS2021_00000_t2.nii.gz')
```

```python
print(flair.shape, flair.affine.shape, flair.get_data_dtype(), flair.header.get_xyzt_units())
print(seg.shape, seg.affine.shape, seg.get_data_dtype(), seg.header.get_xyzt_units())
print(t1.shape, t1.affine.shape, t1.get_data_dtype(), t1.header.get_xyzt_units())
print(t1ce.shape, t1ce.affine.shape, t1ce.get_data_dtype(), t1ce.header.get_xyzt_units())
print(t2.shape, t2.affine.shape, t2.get_data_dtype(), t2.header.get_xyzt_units())
```

```python
data_flair = flair.get_fdata()
data_seg = seg.get_fdata()
data_t1 = t1.get_fdata()
data_t1ce = t1ce.get_fdata()
data_t2 = t2.get_fdata()

print(data_flair.shape, data_seg.shape, data_t1.shape, data_t1ce.shape, data_t2.shape)
```

```python
def plot_slice(slice_num):
    fig, axs = plt.subplots(2, 3, figsize=(15,10))
    axs[0,0].imshow(data_flair[:,:,slice_num], cmap='gray')
    axs[0,1].imshow(data_seg[:,:,slice_num], cmap='gray')
    axs[0,2].imshow(data_t1[:,:,slice_num], cmap='gray')
    axs[1,0].imshow(data_t1ce[:,:,slice_num], cmap='gray')
    axs[1,1].imshow(data_t2[:,:,slice_num], cmap='gray')
    plt.show()
```

```python
for i in range(12):
    plot_slice(i*10)
```
